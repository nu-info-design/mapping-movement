resettlement_df

```python
import pandas as pd

resettlement_df = pd.read_csv("data.csv")
resettlement_df.head()
```

```python
resettlement_df.drop("Year", axis=1, inplace=True, errors='ignore')
resettlement_df.head()
```

```python
# calculate flow_a
flow_a_df = resettlement_df.groupby(["Country of origin", "Country of asylum"])["Total"].sum().reset_index()
flow_a_df.rename(columns={"Total": "flow_a"}, inplace=True)

# merge flow_a_df with resettlement_df
resettlement_df = resettlement_df.merge(flow_a_df, on=["Country of origin", "Country of asylum"])

# rename columns
resettlement_df = resettlement_df.rename(columns={
    "Country of origin": "origin",
    "Country of asylum": "dest_a",
    "flow_a": "flow_a",
    "Country of resettlement": "dest_b",
    "Total": "flow_b"
})

#reorder columns
resettlement_df = resettlement_df[["origin", "dest_a", "flow_a", "dest_b", "flow_b"]]

resettlement_df.head()
```

```python
#renaming countries

# replace "Rep." with "Republic" in the 'origin', 'dest_a', and 'dest_b' columns
resettlement_df['origin'] = resettlement_df['origin'].str.replace('Rep.', 'Republic', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Rep.', 'Republic', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Rep.', 'Republic', regex=False)

# deplace 'Dem.' with 'Democratic'
resettlement_df['origin'] = resettlement_df['origin'].str.replace('Dem.', 'Democratic', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Dem.', 'Democratic', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Dem.', 'Democratic', regex=False)

# remove any text within parentheses (and the parentheses)
resettlement_df['origin'] = resettlement_df['origin'].str.replace(r'\(.*\)', '', regex=True)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace(r'\(.*\)', '', regex=True)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace(r'\(.*\)', '', regex=True)

# deplace 'United Kingdom of Great Britain and Northern Ireland' with 'United Kingdom'
resettlement_df['origin'] = resettlement_df['origin'].str.replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom', regex=False)

#vietnam
resettlement_df['origin'] = resettlement_df['origin'].str.replace('Viet Nam', 'Vietnam', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Viet Nam', 'Vietnam', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Viet Nam', 'Vietnam', regex=False)

#tanzania 
resettlement_df['origin'] = resettlement_df['origin'].str.replace('United Republic of Tanzania', 'Tanzania', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('United Republic of Tanzania', 'Tanzania', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('United Republic of Tanzania', 'Tanzania', regex=False)

#laos
resettlement_df['origin'] = resettlement_df['origin'].str.replace('Lao People\'s Democratic Republic', 'Laos', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Lao People\'s Democratic Republic', 'Laos', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Lao People\'s Democratic Republic', 'Laos', regex=False)

#syria
resettlement_df['origin'] = resettlement_df['origin'].str.replace('Syrian Arab Republic', 'Syria', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Syrian Arab Republic', 'Syria', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Syrian Arab Republic', 'Syria', regex=False)

resettlement_df['origin'] = resettlement_df['origin'].str.replace('Democratic Republic of the Congo', 'DR Congo', regex=False)
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.replace('Democratic Republic of the Congo', 'DR Congo', regex=False)
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.replace('Democratic Republic of the Congo', 'DR Congo', regex=False)


# check countries
print(pd.concat([resettlement_df['origin'], resettlement_df['dest_a'], resettlement_df['dest_b']]).unique())
```

```python
#strip leading and trailing spaces from column names
resettlement_df.columns = resettlement_df.columns.str.strip()

# strip leading and trailing spaces from values in the 'origin', 'dest_a', and 'dest_b' columns
resettlement_df['origin'] = resettlement_df['origin'].str.strip()
resettlement_df['dest_a'] = resettlement_df['dest_a'].str.strip()
resettlement_df['dest_b'] = resettlement_df['dest_b'].str.strip()

# make sure the flow columns are numeric
resettlement_df['flow_a'] = pd.to_numeric(resettlement_df['flow_a'], errors='coerce')
resettlement_df['flow_b'] = pd.to_numeric(resettlement_df['flow_b'], errors='coerce')

# remove rows with any missing or NaN values
resettlement_df.dropna(inplace=True)

# reset the index
resettlement_df.reset_index(drop=True, inplace=True)

# check countries
print(pd.concat([resettlement_df['origin'], resettlement_df['dest_a'], resettlement_df['dest_b']]).unique())
```

```python
resettlement_df.to_csv('data/resettlement_df.csv', index=False)
print("DataFrame saved to 'data/resettlement_df.csv'")
```

circle scaling df

```python
# total_movement_df for circle scaling

# sum of 'flow_b' movements for each country (origin, dest_a, and dest_b)
total_movement_origin = resettlement_df.groupby('origin')['flow_b'].sum().reset_index()
total_movement_dest_a = resettlement_df.groupby('dest_a')['flow_b'].sum().reset_index()
total_movement_dest_b = resettlement_df.groupby('dest_b')['flow_b'].sum().reset_index()

# combine all country columns (origin, dest_a, dest_b) into one
total_movement_origin.columns = ['country', 'total_movement']
total_movement_dest_a.columns = ['country', 'total_movement']
total_movement_dest_b.columns = ['country', 'total_movement']

# combine total movement data from each column
total_movement_df = pd.concat([total_movement_origin, total_movement_dest_a, total_movement_dest_b])

# group by 'Country' and sum the 'Total_movement' to make sure we aggregate the data correctly
total_movement_df = total_movement_df.groupby('country')['total_movement'].sum().reset_index()
total_movement_df
```

```python
import vizutils as vu
```

```python
# define the range for scaling
min_radius = 100  # Minimum radius size
max_radius = 800  # Maximum radius size

# Apply the scaling function
circles_scaled_df = vu.circle_area_scale(
    total_movement_df,                          # The DataFrame
    'total_movement',                            # Column to scale
    total_movement_df['total_movement'].min(),  # Minimum value in the column
    total_movement_df['total_movement'].max(),  # Maximum value in the column
    min_radius,                                 # Min radius for scaled values
    max_radius                                  # Max radius for scaled values
)

circles_scaled_df
```

```python
circles_scaled_df.to_csv('data/circles_scaled_df.csv', index=False)
print("DataFrame saved to 'data/circles_scaled_df.csv'")
```
